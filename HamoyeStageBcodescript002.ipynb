#### QUESTION NUMBER 12

```python
#calculate the linear regression of T2 and T6

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X = energydf[["T2"]]
y = energydf["T6"]

linear_model = LinearRegression()
linear_model.fit(X, y)

predictions = linear_model.predict(X)
round(r2_score(y, predictions), 2)
```

```python
x1 = energydf[["T2"]]
y = energydf["T6"]

x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

#### QUESTION NUMBER 13 TO 17

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error

X = energydf[energydf.columns.difference(["date", "lights", "Appliances"])]
y = energydf["Appliances"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

multiple_linear_model = LinearRegression()
multiple_linear_model.fit(X_train, y_train)

test_predictions = multiple_linear_model.predict(X_test)
```

```python
x2 =  energydf[energydf.columns.difference(["date", "lights", "Appliances"])]
y = energydf["Appliances"]

x = sm.add_constant(x2)
results = sm.OLS(y, x).fit()
results.summary()
```

#### QUESTION 13(specifically)

```python
#Calculating the Mean Absolute Error

round(mean_absolute_error(y_test, test_predictions), 2)
```

#### QUESTION 14(specifically)

```python
#Residual Sum of Squares to measure the level of variance in the error term

round(np.sum(np.square(y_test - test_predictions)) , 2)
```

#### QUESTION 15(specifically)

```python
#Calculating the Root Mean Squared Error

round(mean_squared_error(y_test, test_predictions, squared=False), 3)
```

#### QUESTION 16(specifically)

```python
#Rsquared to evaluate the performance and accuracy of the linear regression model

round(r2_score(y_test, test_predictions), 2)
```

```python
tmptdf = pd.DataFrame([X.columns, multiple_linear_model.coef_]).T
tmptdf.columns = ["attributes", "size"]
tmptdf.sort_values(by="size", ascending=False)
```
